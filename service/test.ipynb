```python
import base64
import requests
```

```python
def image_to_base64(image_path):
    with open(image_path, 'rb') as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return encoded_string
```

```python
image_path = '/app/raw_images_old/test.jpg'
image_base64 = image_to_base64(image_path)
```

```python
endpoint = 'http://localhost:8648/api/face_beautify/'
data = {
    'image': image_base64,
}
response = requests.post(endpoint, json=data)
```

```python
def plot_image(base64_string):
    import io
    import matplotlib.pyplot as plt
    from PIL import Image

    image = Image.open(io.BytesIO(base64.b64decode(base64_string)))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
plot_image(image_base64)
```

```python
plot_image(response.json()['image'])
```
